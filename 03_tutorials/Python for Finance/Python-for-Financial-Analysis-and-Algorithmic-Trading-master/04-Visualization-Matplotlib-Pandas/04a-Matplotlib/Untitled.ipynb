```python
import numpy as np
x = np.arange(0,100)
y = x * 2
z = x ** 2

fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.4,.4])
ax1.plot(x,z)
ax2.plot(x,y)
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_title("t")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.set_title("t")
ax1.set_xlim([0,100])
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

```python
fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(12,10))
ax[0].plot(x,y,"g--",lw=2)
ax[1].plot(x,z,"r-.",lw=3)

plt.tight_layout()
```
